## Darbs ar sarakstiem (Lists)

Saraksti ir viena no svarīgākajām datu struktūrām Python.


```python
# 1. Izveidojam sarakstu ar amplitūdas mērījumiem (metros)
amplitudes = [0.50, 0.42, 0.35, 0.29, 0.24, 0.20]

# 2. Pievienojam jaunu mērījumu (nākamais cikls)
amplitudes.append(0.16)

# 3. Aprēķinām amplitūdas samazinājumu starp pirmo un pēdējo mērījumu
zudums = amplitudes[0] - amplitudes[-1]

# 4. Sarakstu funkcijas inženiertehniskai analīzei
print(f"Kopā reģistrēti {len(amplitudes)} cikli.")
print(f"Sākotnējā amplitūda: {max(amplitudes)} m")
print(f"Beigu amplitūda: {min(amplitudes)} m")
print(f"Kopējais amplitūdas kritums: {zudums:.2f} m")
```

## Vienkārša funkcija un loģika

Funkcijas palīdz automatizēt darbības, kuras mēs mēdzam atkārtot.

```python
def parbaudit_skaitli(n):
    if n > 0:
        return "Skaitlis ir pozitīvs"
    elif n < 0:
        return "Skaitlis ir negatīvs"
    else:
        return "Tā ir nulle"

# Izmēģini mainīt skaitli šeit:
tests = -5
print(f"Pārbaude: {tests} -> {parbaudit_skaitli(tests)}")
```

## Vizualizācija

```python
import matplotlib.pyplot as plt
import numpy as np

# Definējam laika asi (0 līdz 2 periodiem)
t = np.linspace(0, 4 * np.pi, 100)
# Harmonisko svārstību vienādojums: x(t) = A * cos(omega * t)
x = 2 * np.cos(2.5 * t)

plt.figure(figsize=(10, 4))
plt.plot(t, x, label='Atsperes nobīde')
plt.title('Harmoniskās svārstības')
plt.xlabel('Laiks (s)')
plt.ylabel('Amplitūda')
plt.grid(True)
plt.legend()
plt.show()
```
